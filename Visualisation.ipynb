# MATPLOTLIB

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

Line Chart


```python
x=[10,14,18,22]
y=[45,67,89,55]
fig=plt.figure(figsize=(5,5))
plt.plot(x,y,marker='o',c='purple')
plt.title('Line plot')
plt.xlabel('x_axis',fontsize=14,color='purple')
plt.ylabel('y_axis',fontsize=14,color='purple')
plt.grid()
```

```python
#for two variable
plt.figure(figsize=(7,7))
x=[1,2,3,4,5,6]
y=[100,234,311,344,367,875]
z=[211,900,567,542,222,777]
plt.plot(x,y,marker='D',c='purple',label='Noida')
plt.plot(x,z,marker='D',c='orange',label='Gurgao')
plt.grid()
plt.legend()
plt.title('sales vs month')
plt.xlabel('months',fontsize=17,color='Purple')
plt.ylabel('sales',fontsize=17,color='Purple')
```

BAR CHART

```python
plt.figure(figsize=(5,5))
plt.bar(x,z,label='chennai')
plt.title('sales vs month of chennai',fontsize=25)
plt.xlabel('month',fontsize=20,color='Black')
plt.ylabel('sales',fontsize=20,color='Black')
plt.show()
```

Pie-Chart

```python
x=['Data analyst','Data scientist','MACHINE LEARNING ENGG']
Y=[100000,200000,160000]
plt.pie(Y,labels=x)
plt.show()
```

BOXPLOT

```python
a=np.linspace(1,10,100)
plt.boxplot(a)
plt.show()
```

Histogram

```python
l=['a','a','k','k','k','e','p']
plt.hist(l)

plt.title('Histogram',fontsize=25)
plt.xlabel('alphabets',fontsize=20)
plt.ylabel('counts',fontsize=20)

plt.show()
```

Scatterplot

```python
x=np.random.randint(0,100,30)
y=np.random.randint(100,1000,30)
z=np.sqrt(y)
plt.scatter(x,y,marker='D')
```

SUBPLOTS

```python
#MULTIPLE PLOT
##SUBPLOT
x=np.linspace(1,10,100)
plt.figure()
plt.subplot(221)#2rows 1 column 1st grapth
plt.plot(x,np.sin(x))
plt.title('sin')
plt.subplot(222)
plt.plot(x,np.cos(x))#2rows 1 col and 2nd graph
plt.title('cos')
plt.subplot(223)
plt.plot(x,np.exp(x))
plt.title('exp')
plt.subplot(224)
plt.plot(x,np.log(x))
plt.title('log')
```

```python
x=np.random.randint(1,100,10)
y=np.random.randint(20,30,10)
```

```python
plt.figure()
plt.subplot(221)
plt.scatter(x,y)
plt.subplot(222)
plt.plot(y)
plt.subplot(223)
plt.bar(x,y)
plt.subplot(224)
plt.boxplot(y)
```
